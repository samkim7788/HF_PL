```python
import pandas as pd

# df["dialogue"] 안에 [{'user_id': 0, 'utter': '...'}, ...] 리스트가 들어있다고 가정
def extract_user_utters(df, dialogue_col="dialogue", user_id=0):
    out = []
    for conv in df[dialogue_col]:
        # conv: list[dict]
        for turn in conv:
            if turn.get("user_id") == user_id:
                utt = (turn.get("utter") or "").strip()
                if utt:
                    out.append(utt)
    return pd.Series(out, name="text")

texts = extract_user_utters(df)       # 모든 사용자 문장(개수 상관없음)
print(len(texts), "문장 추출 완료")
print(texts.head())
```

```python
# DataFrame → 리스트 변환 (문자열로 보장)
X_train = train_df['situation'].astype(str).tolist()
y_train = train_df['user_utter'].astype(str).tolist()

X_test = test_df['situation'].astype(str).tolist()
y_test = test_df['user_utter'].astype(str).tolist()
```

```python
from transformers import T5Tokenizer, TFT5ForConditionalGeneration

# 토크나이저 불러오기
tokenizer = T5Tokenizer.from_pretrained("t5-small")

# 데이터 토큰화 함수
def tokenize(X, y, max_length=128):
    model_inputs = tokenizer(
        X,
        max_length=max_length,
        truncation=True,
        padding="max_length",
        return_tensors="tf"
    )

    labels = tokenizer(
        y,
        max_length=max_length,
        truncation=True,
        padding="max_length",
        return_tensors="tf"
    )["input_ids"]

    model_inputs["labels"] = labels
    return model_inputs
```

```python
# 데이터셋 만들기
train_encodings = tokenize(X_train, y_train)
test_encodings = tokenize(X_test, y_test)

train_dataset = tf.data.Dataset.from_tensor_slices((
    {
        "input_ids": train_encodings["input_ids"],
        "attention_mask": train_encodings["attention_mask"],
        "labels": train_encodings["labels"],
    }
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    {
        "input_ids": test_encodings["input_ids"],
        "attention_mask": test_encodings["attention_mask"],
        "labels": test_encodings["labels"],
    }
))


# 배치 사이즈 적용
train_dataset = train_dataset.shuffle(1000).batch(16)
test_dataset = test_dataset.batch(16)
```

```python
from transformers import create_optimizer
import tensorflow as tf

num_epochs = 3
num_train_steps = len(train_dataset) * num_epochs
num_warmup_steps = int(num_train_steps * 0.1)

optimizer, lr_schedule = create_optimizer(
    init_lr=5e-5,
    num_train_steps=num_train_steps,
    num_warmup_steps=num_warmup_steps,
    weight_decay_rate=0.01
)

# 👇 핵심 수정: loss를 직접 지정 (문자열 → int라서 sparse categorical crossentropy 사용)
model.compile(
    optimizer=optimizer,
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy']
)

history = model.fit(
    train_dataset,
    validation_data=test_dataset,
    epochs=num_epochs
)
```
