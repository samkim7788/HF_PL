```python
from datasets import load_dataset
from google.colab import userdata

# Colab 시크릿에서 토큰 불러오기
token_data = userdata.get('use_token')

from datasets import load_dataset

dataset = load_dataset("passing2961/KorEmpatheticDialogues")

# pandas DataFrame 변환
df = dataset["train"].to_pandas()
print(df.head())
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder

# 예: df["emotion"] 이 문자열 라벨이라고 가정
encoder = LabelEncoder()
df["emotion_label"] = encoder.fit_transform(df["emotion"])

# TensorFlow Dataset 만들 때는 emotion_label 사용
y = df["emotion_label"].astype("int32")
```

```python
# 데이터 train, test 분류
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

print("Train:", train_df.shape, "Test:", test_df.shape)
```

```python
train_df =train_df.dropna(how='any')
test_df = test_df.dropna(how='any')

display(train_df.info())
display(test_df.info())
```

```python
X_train = train_df['situation'].values.tolist()
y_train = train_df['emotion_label'].values.tolist()

X_test = test_df['situation'].values.tolist()
y_test = test_df['emotion_label'].values.tolist()
```

# 모델 학습

```python
from transformers import AutoTokenizer,TFAutoModelForSequenceClassification

model = TFAutoModelForSequenceClassification.from_pretrained("monologg/kobert", from_pt=True, num_labels=32)
tokenizer = AutoTokenizer.from_pretrained("monologg/kobert")
```

```python
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

```python
X_train = tokenizer(X_train, padding=True, truncation=True)
X_test = tokenizer(X_test, padding=True, truncation=True)

X_train[:3], X_test[:3]
```

```python
print(X_train['input_ids'][0]) # 패딩처리 됨
print(X_train['attention_mask'][0])
print(X_train['token_type_ids'][0])
```

```python
type(X_train)
```

```python
import tensorflow as tf

train_ds = tf.data.Dataset.from_tensor_slices((dict(X_train), y_train)) # 딕셔너리로 변환
test_ds = tf.data.Dataset.from_tensor_slices((dict(X_test), y_test))
```

```python
train_dataset = train_ds.shuffle(10000).batch(8).prefetch(tf.data.AUTOTUNE) # 병렬처리 가능하도록 세팅
test_dataset = test_ds.batch(8).prefetch(tf.data.AUTOTUNE)
```

```python
from transformers import TFAutoModelForSequenceClassification

model = TFAutoModelForSequenceClassification.from_pretrained(
    "monologg/kobert",
    num_labels=32,
    from_pt=True
)
```

```python
from transformers import create_optimizer
from tensorflow.keras.losses import SparseCategoricalCrossentropy

num_train_steps = len(train_dataset) * 5
num_warmup_steps = int(num_train_steps * 0.1)

optimizer, _ = create_optimizer(
    init_lr=5e-5,
    num_train_steps=num_train_steps,
    num_warmup_steps=num_warmup_steps,
    weight_decay_rate=0.1
)

loss = SparseCategoricalCrossentropy(from_logits=True)

model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])

model.fit(train_dataset, epochs=5, validation_data=test_dataset)
```

```python
model.save_pretrained('monkey777/cheer-up-all')
tokenizer.save_pretrained('monkey777/cheer-up-all')
```

```python
from transformers import TextClassificationPipeline

sentiment_classifier = TextClassificationPipeline(
    tokenizer=tokenizer,
    model=model,
    framework='tf',
    return_all_scores=True
)
```

```python
sentiment_classifier('감기에 걸렸는데 어떻하죠???')
```
